## Subplots and Combining Visualizations ##

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# Data set
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
# Set axis (scaffolding)
fig,(ax1, ax2) = plt.subplots(ncols=2,
nrows=1,
figsize=(7, 3))
```

```python
# Plot
fig, (ax1, ax2) = plt.subplots(ncols=2,
nrows=1,
figsize=(7, 3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
fig.show()
```

```python
# Modify plot
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(7, 3))

# ---- Scatter plot ----
ax1.scatter(x1, y1, color="blue", label="Data")

# Annotation
ax1.annotate(
    "Important point",
    xy=(x1[0], y1[0]),
    xytext=(x1[0] + 1, y1[0] + 1),
    arrowprops=dict(arrowstyle="->")
)

# Scale + labels
ax1.set_xlabel("x1")
ax1.set_ylabel("y1")
ax1.set_xlim(min(x1), max(x1))
ax1.set_ylim(min(y1), max(y1))

# Legend
ax1.legend()

# ---- Bar plot ----
ax2.bar(x2, y2, color="green", label="Counts")

# Annotation
ax2.annotate(
    "Highest bar",
    xy=(x2[0], y2[0]),
    xytext=(x2[0], y2[0] + 1),
    arrowprops=dict(arrowstyle="->")
)

# Scale + labels
ax2.set_xlabel("x2")
ax2.set_ylabel("y2")
ax2.set_ylim(0, max(y2) + 1)

# Legend
ax2.legend()

plt.tight_layout()
plt.show()
```

**Adjusting layouts**

```python
# Let’s try adding some very large x axis titles to our previous plot
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
['ax2', 'ax3']],
figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
```

• We can use layouts to make sure that our subplots fit neatly in our figure area

• There are two main kinds of layout we consider:

         ◦ Tight layout adjusts subplots so tick labels, axis labels, and titles don't overlap or leave the figure area
         
         ◦ Constrained layout works similarly except it also fits things like legends or colorbars

```python
# Constrained layout
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
['ax2', 'ax3']],
figsize=(7, 4),
layout = "constrained")
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label', fontsize=18)
someaxes["ax2"].set_xlabel('Another Label', fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
```

**Multiple viz on one axes**

```python
# first make our sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])
# define our figure and axes (just one this time)
fig, ax = plt.subplots(figsize=(7, 3))
# now call both bar and plot elements to the same axes (ax)
ax.bar(x, y1,
color = "indigo")
ax.plot(x, y2,
color = "red")
```

**Add error information**

```python
# First, calculate standard deviation of our data
y2_sd = np.std(y2)

# Then plot our line as before
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "blue")

# Add error bars
ax.errorbar(x, #our x values
y2, #our y values
yerr = y2_sd,
color = 'red',
fmt = "none") # excludes data points
```

Customizing errorbar appearance


```python
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2,
color = "red")
ax.errorbar(x,
y2,
yerr = y2_sd,
fmt = "none",
ecolor= "indigo",
elinewidth= 4,
capsize = 6,
capthick= 4 )
```

If we don’t want to see error bars for every single point, we can specify intervals using errorevery

```python
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2,
color = "red")

ax.errorbar(x,
y2,
yerr = y2_sd,
fmt = "none",
ecolor = "indigo",
elinewidth = 4,
capsize = 6,
capthick = 4,
errorevery= 2 )
```
