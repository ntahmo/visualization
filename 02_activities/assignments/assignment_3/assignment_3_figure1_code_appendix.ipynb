```python
# Set up: load libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load LRT delay data: https://open.toronto.ca/dataset/ttc-lrt-delay-data/
LRT_LRT_delays = pd.read_csv("C:/Users/nancy/Downloads/TTC LRT LRT_delays.csv")

print(LRT_LRT_delays.columns)
```

```python
# Combine Date + Time into a datetime, then extract hour
# (This assumes Date and Time are strings like "2024-01-15" and "13:45")
LRT_delays["datetime"] = pd.to_datetime(LRT_delays["Date"].astype(str) + " " + LRT_delays["Time"].astype(str),
                                    errors="coerce")
LRT_delays["hour"] = LRT_delays["datetime"].dt.hour

# 3) Drop rows missing essentials for this plot
plot_df = LRT_delays.dropna(subset=["hour", "Line", "Min Delay"])

# 4) Summarize: mean delay by hour within each line
summary = (
    plot_df.groupby(["Line", "hour"])["Min Delay"]
    .mean()
    .reset_index()
)

```

```python
# 5) Plot (matplotlib “fig, ax” pattern)
plt.style.use("default")  # you can try other styles like 'fivethirtyeight'
fig, ax = plt.subplots(figsize=(7, 4))

# Plot one line per LRT line
lines = summary["Line"].unique()

for ln in lines:
    temp = summary[summary["Line"] == ln]
    ax.plot(temp["hour"], temp["Min Delay"], marker="o", linewidth=2, label=str(ln))

# Labels + title
ax.set_title("Average TTC LRT Delay by Hour (minutes)")
ax.set_xlabel("Hour of day")
ax.set_ylabel("Average delay (minutes)")

# Make ticks readable and consistent
ax.set_xticks(np.arange(0, 24, 2))
ax.grid(axis="y", linewidth=1, linestyle="--", alpha=0.5)

# Legend (keep it readable)
ax.legend(loc="upper left", frameon=True)

# Layout so labels don’t get cut off
fig.tight_layout()

plt.show()
```
